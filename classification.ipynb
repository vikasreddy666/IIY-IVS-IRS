```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

text_data = [
    "free ticket win",
    "good morning",
    "earn money online",
    "claim the prize"
]

labels = [1, 0, 1, 0]  # 1 = spam, 0 = normal

data = pd.DataFrame({
    "msg": text_data,
    "target": labels
})

print("Dataset Used:")
print(data)

vector = CountVectorizer()
x = vector.fit_transform(data["msg"])
y = data["target"]

lr = LogisticRegression()
lr.fit(x, y)

test = ["win free ticket"]
test_vec = vector.transform(test)

prediction = lr.predict(test_vec)

print("\nTesting Message:", test[0])

if prediction[0] == 1:
    print("Prediction: Spam Message")
else:
    print("Prediction: Normal Message")
```
